```python
import gensim
from gensim.corpora import Dictionary
import os
from sklearn.manifold import TSNE
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class MySentences(object):
    def __init__(self, path):
        self.path = path
        self.errors = []

    def __iter__(self):
        for f in os.listdir(self.path):
            with open(os.path.join(self.path,f), encoding="utf8") as t:
                text = t.read()
                text = [ i.strip('\n') for i in text.split(" ")]
                yield text
sentences = MySentences('texts') # a memory-friendly iterator
```

```python
model = gensim.models.Word2Vec(sentences, min_count=50, size = 200)
```

```python
model.most_similar("nautae")
```
